Импортируем библиотеки

```python
import pandas as pd
```

Читаем данные с Таблицы-excel

```python
df = pd.read_excel('tabl1.xlsx')
```

Поменяем слегка Датафрейм, уберем лишние столбцы и добавим чистую прибыль между продажей и закупками.

```python
df.drop(["Код товара", "Ед.изм."], axis=1, inplace=True)
df["Прибыль"] = df["Сумма в ценах продажи"] - df["Сумма в ценах закупки"]
```

Теперь смотрим на список агентов.

```python
list_agents = df["Агент"].drop_duplicates()
list_agents
```

Смотрим на общий отчет

```python
df["Прибыль"].describe(include='all')
```

Групируем таблицу с агентами и описательной аналитики

```python
descriptive_stats = df.groupby('Агент')['Прибыль'].agg(['count', 'mean', 'std', 'min', 'max', 'median', 'sum'])
```

```python
descriptive_stats.sort_values(by='sum', ascending=False)
```

Вывод.

Лучшие агенты - это агент20, агент22, агент9, так как занимают топ-3 агентов по прибыли.
